```python
x=var('x')
y=function('y')(x)
eqd=diff(y,x)+(1/x)*y==x^3
desolve(eqd,[y,x])
```

```python
sol=desolve(eqd,[y,x])
_C=var('_C')
sol.substitute(_C==1)
```

```python
sol1=sol.substitute(_C==1)
```

```python
plot(sol1,x,-2,2, ymin = -10, ymax = 10)
```
