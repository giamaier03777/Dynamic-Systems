```python
t,k,x0=var('t,k, x0')
x=function('x')(t)
eqd=diff(x,t)==-k*x
desolve(eqd,[x,t],ics=[0,x0])
```

```python
sol(t,k,x0)=desolve(eqd,[x,t],ics=[0,x0])
sol(t,k,x0)
```

```python
plot([sol(t,-1, 10),sol(t,1, 30),sol(t,2, 20)],t,-2,2)
```

```python
t,k,Tu,T0=var('t,k,Tu,T0')
T=function('T')(t)
eqd=diff(T,t)==-k*(T-Tu)
desolve(eqd,[T,t],ics=[0,T0])
```

```python
sol(t,k,Tu,T0)=desolve(eqd,[T,t],ics=[0,T0])
sol(t,k,Tu,T0)
```

```python
plot([sol(t,-1,20,30),sol(t,1,30,10),sol(t,2,15, 25)],t,-2,2)
```
